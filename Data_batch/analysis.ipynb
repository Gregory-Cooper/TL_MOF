```python
import pandas as pd
holder=[1,8,16,32,64,128,256]
dfs=[]
for i in holder:
    temp=i
    base=f"data_epochs-10000_bs-{str(temp)}.json"
    dfs.append(pd.read_json(base))

dfs[0]['train_loss'][:].ewm(span = 500).mean().plot()
```

```python
for g in dfs[1].columns:
    for count,i in enumerate(dfs):
        if count==0:
            ax=i[[g]][500:].ewm(span = 100).mean().plot()
        i[[g]][500:].ewm(span = 500).mean().plot(ax=ax)
```
