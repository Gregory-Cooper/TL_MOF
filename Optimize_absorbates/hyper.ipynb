```python
def cat_ratio(x):
    if x<.2:
        return "<.2"
    elif x<.4:
        return ".2-.4"
    elif x<.6:
        return ".4-.6"
    elif x<.8:
        return ".6-.8"
    else:
        return ">.8"
hold["Ratio"]=hold["Ratio"].apply(cat_ratio)
```
