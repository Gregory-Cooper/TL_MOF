```python
import numpy as np
Start=100
game=True
round=0
number=[]
for i in range(10000):
    while game:
        round+=1
        Start=sum(np.random.choice(2,Start,p=[1/3,2/3]))
        if Start==0:
            game=False
    game=True
    number.append(round)
    round=0
    Start=100
```

```python
import matplotlib.pyplot as plt
number=np.array(number)
print("The average number of rounds is: ", number.mean(), "or ", int(number.mean()))
plt.hist(number,bins=25,density=True)
x=np.linspace(1,number.max(),number.max()+1)
p=2/3
y=(1-p**x)**100-(1-p**(x-1))**100
plt.plot(x,y)
```
